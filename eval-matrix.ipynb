```python
I1_incptn_lstm ='a zebra standing in a field with a zebra'
I1_incptn_trfmr = 'a zebra standing in a field with a field'
I1_xcptn_lstm = 'a man is standing on a skateboard in the snow'
I1_xcptn_trfmr = 'a giraffe standing in a field with a field'
I1_irn_lstm = 'a zebra standing in a field with a zebra'
I1_irn_trfmr = 'a zebra standing in a field with a field'
I1_vgg16_lstm = 'a man is riding a skateboard on a skateboard'
I1_vgg16_trfmr = 'a group of people standing on a field'
```

```python
I1_incptn_lstm_tokens = nltk.word_tokenize(I1_incptn_lstm.lower())
I1_incptn_trfmr_tokens = nltk.word_tokenize(I1_incptn_trfmr.lower())
I1_xcptn_lstm_tokens = nltk.word_tokenize(I1_xcptn_lstm.lower())
I1_xcptn_trfmr_tokens = nltk.word_tokenize(I1_xcptn_trfmr.lower())
I1_irn_lstm_tokens = nltk.word_tokenize(I1_irn_lstm.lower())
I1_irn_trfmr_tokens = nltk.word_tokenize(I1_irn_trfmr.lower())
I1_vgg16_lstm_tokens = nltk.word_tokenize(I1_vgg16_lstm.lower())
I1_vgg16_trfmr_tokens = nltk.word_tokenize(I1_vgg16_trfmr.lower())
```

```python
ref_1_tokens = nltk.word_tokenize(ref_1.lower())
ref_2_tokens = nltk.word_tokenize(ref_2.lower())
```

```python
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_incptn_lstm_tokens)
print("BLEU score for incptn_lstm:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_incptn_trfmr_tokens)
print("BLEU score for incptn_trfmr:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_xcptn_lstm_tokens)
print("BLEU score for xcptn_lstm:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_xcptn_trfmr_tokens)
print("BLEU score for xcptn_trfmr:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_irn_lstm_tokens)
print("BLEU score for irn_lstm:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_irn_trfmr_tokens)
print("BLEU score for irn_trfmr:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_vgg16_lstm_tokens)
print("BLEU score for vgg16_lstm:", bleu_score)
bleu_score = sentence_bleu([ref_1_tokens,ref_2_tokens], I1_vgg16_trfmr_tokens)
print("BLEU score for vgg16_trfmr:", bleu_score)
```
